```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=pd.read_csv('car data.csv')
```

```python
df = df.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}})
df = df.replace({'Selling_type':{'Dealer':0,'Individual':1}})
df = df.replace({'Transmission':{'Manual':0,'Automatic':1}})
```

```python
X=df.drop(['Car_Name','Selling_Price'],axis=1)
y = df[['Selling_Price']]
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.8,random_state=3)
```

```python
features = X.columns   # all 6 features

for col in features:
    sns.scatterplot(x=df[col], y=df['Selling_Price'])
    plt.title(f'{col} vs Selling_Price')
    plt.show()
```

```python
Pred_model=LinearRegression()
```

```python
Pred_model.fit(X_train,y_train)
```

```python
prediction=Pred_model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, prediction)
r2 = r2_score(y_test, prediction)

print(f"Mean Squared Error : {mse}")
print(f"R2 Score : {r2*100:.2f}")
```
